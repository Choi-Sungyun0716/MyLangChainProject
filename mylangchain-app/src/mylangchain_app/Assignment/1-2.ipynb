```python
from dotenv import load_dotenv
import os
# .env 파일을 불러와서 환경 변수로 설정
load_dotenv(dotenv_path='../.env') # 주소는 안줘도 됨

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
print(OPENAI_API_KEY[:10])
```

```python

from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from pprint import pprint

llm = ChatOpenAI(
    api_key=OPENAI_API_KEY,
    base_url="https://api.groq.com/openai/v1",  # Groq API 엔드포인트
    #model="meta-llama/llama-4-scout-17b-16e-instruct",  # Spring AI와 동일한 모델
    model="openai/gpt-oss-120b",
    temperature=0.7
)

prompt_step1 = ChatPromptTemplate.from_messages([
    ("system", "너는 여행 가이드 AI야."),
    ("user", "{location}으로 여행을 간다면 꼭 가야 할 대표적인 명소 1가지만 추천해줘. 명소 이름만 답변해.")
])

chain_step1 = prompt_step1 | llm | StrOutputParser()


prompt_step2 = ChatPromptTemplate.from_messages([
    ("system", "너는 여행 가이드 AI야."),
    ("user", "추천받은 명소는 {landmark}야. 이 명소의 역사, 특징, 방문 팁을 자세히 알려줘.")
])

chain_step2 = prompt_step2 | llm | StrOutputParser()

def travel_info_system(location: str):
   
    landmark = chain_step1.invoke({"location": location})
    print("📍 1단계 추천 명소:", landmark)

   
    details = chain_step2.invoke({"landmark": landmark})
    print("\n📖 2단계 명소 상세 정보:\n", details)

    return {"landmark": landmark, "details": details}


response = travel_info_system("로마")
pprint(response)
```
